```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
data = {

"text":[

"Invoice Number INV001 Amount 25000 GST Payment Due",
"Invoice Total Tax Customer Billing Address",
"Invoice Amount Paid Bank Details",

"Purchase Order PO123 Supplier Delivery Date",
"Purchase Order Vendor Material Required",
"PO Number Quantity Unit Price",

"Mill Test Certificate ASTM A106 Grade B Heat Number",
"MTC Chemical Composition Mechanical Properties",
"Mill Test Certificate Material Inspection",

"Engineering Drawing Revision A Scale 1:10",
"AutoCAD Drawing Dimension Sheet",
"Technical Drawing Assembly Layout",

"Project Report Introduction Methodology Conclusion",
"Final Report Analysis Recommendation",
"Inspection Report Observation Summary"

],

"label":[

"Invoice",
"Invoice",
"Invoice",

"Purchase Order",
"Purchase Order",
"Purchase Order",

"Mill Test Certificate",
"Mill Test Certificate",
"Mill Test Certificate",

"Drawing",
"Drawing",
"Drawing",

"Report",
"Report",
"Report"

]

}

df = pd.DataFrame(data)
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df["text"],
    df["label"],
    test_size=0.2,
    random_state=42
)
```

```python
model = Pipeline([

    ("tfidf", TfidfVectorizer()),

    ("classifier", LogisticRegression(max_iter=1000))

])
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)

print(classification_report(y_test, predictions))
```

```python
new_doc = """

Invoice Number INV987
Customer Name ABC Pvt Ltd
GST Amount 4500
Grand Total 78000
Payment Due 30 Days

"""

prediction = model.predict([new_doc])[0]

prob = model.predict_proba([new_doc]).max()

print("Predicted Class :", prediction)
print("Confidence :", round(prob*100,2),"%")
```
